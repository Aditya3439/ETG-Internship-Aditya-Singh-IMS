# Reading the Inventory

```python
import json
fd = open("items.json",'r')
r = fd.read()
fd.close()
records = json.loads(r)
```

```python
records
```

# Purchasing an item from the Inventory

```python
n = int(input("Enter Total number of items you want to Purchase: "))
sum = 0
def Merge(dict1, dict2):
    return(dict2.update(dict1))
tsale={}
print("************************************************************************")
while n!=0:   
    user_prod = str(input("Enter the Product's ID you want to Purchase: "))
    user_quantity = int(input("Enter the Qunatity you need: "))
    print("You Purchased: ", records[user_prod]["name"])
    print("Quantity: ", user_quantity, "Price: ", records[user_prod]["price"])
    sum = sum + (records[user_prod]["price"] * user_quantity)
    records[user_prod]["quantity"] = records[user_prod]["quantity"] - user_quantity
    sales = {1: {"prod": user_prod, "name": records[user_prod]["name"], "quantity": user_quantity, "amount": records[user_prod]["price"] * user_quantity}}
    Merge(sales, tsale)
    n -= 1
print("-------------------------------------------------------------------------")
print("Your Total Billing Amount is: ", sum)
print("Thanks for Visting our Inventory!!!!!!!")
print("Do Come Next Time Also!!!!!!!!!!")
print("*************************************************************************")
    
   
```
